```python
import numpy as np
marks = np.array([
    #s1  s2  s3  s4  s5
    [40, 50, 60, 70, 80], # mathematics
    [95, 97, 65, 98, 90], # physics
    [33, 45, 67, 72, 89], # chemistry
])
marks.shape
avg_student = marks.mean(axis=0)
print("average of each student", avg_student)

max_score = marks.max(axis=1)
print("maximum marks in each subject", max_score)

extraMarks = marks + 5
print("updated score", extraMarks)
```
